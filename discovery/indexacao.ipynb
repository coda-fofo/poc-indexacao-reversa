```python
# Lendo o arquivo CSV com os tatuadores
import pandas as pd

df = pd.read_csv('./assets/tatuadores.csv', sep=';')
df.head()
```

```python
# Gerando o SQL pra inserir no banco!
for i, row in df.iterrows():
    nome_de_usuario = row['nome_de_usuario']
    hashtags = ", ".join(map(lambda hashtag: f"'{hashtag}'" , row['hashtags'].split(',')))

    print('INSERT INTO Tatuador_Hashtag (tatuador_id, hashtag_id)')
    print('SELECT t.id, h.id FROM Tatuador t')
    print('JOIN (SELECT id, hashtag FROM Hashtag) h')
    print(f'ON h.hashtag IN ({hashtags})')
    print(f"WHERE t.nome_de_usuario = '{nome_de_usuario}';")
    print()
```

```python
# Bora partir para a indexação
def transformar_em_frase(row) -> str:
    return row['nome_de_usuario'] + ' ' + row['nome_completo'] + ' ' + row['descricao'] + ' ' + row['hashtags'].replace(',', ' ')

frases = [transformar_em_frase(coluna) for _, coluna in df.iterrows()]
frases
```

```python
from requests import post

frases_tokenizadas = []

for frase in frases:
    response = post('http://localhost:8010/pipeline/tokenizar', params={ 'versao': 'tokenizador-v1' }, json={ 'frase': frase })
    frases_tokenizadas.append(response.json()['tokens'])

frases_tokenizadas
```

```python
from collections import defaultdict

inverted_index = defaultdict(set)

for idx, tokenizado in enumerate(frases_tokenizadas):
    for token in tokenizado:
        inverted_index[token].add(idx + 1)
        
df_index = pd.DataFrame([(term, list(doc_ids)) for term, doc_ids in inverted_index.items()], columns=["Term", "Documents"])
df_index
```

```python
# Bora inserir no índice!

print('INSERT INTO TatuadorIndice (token, tatuadores_ids)')
print('VALUES')

for index, row in df_index.iterrows():
    token = row['Term']
    documents = ", ".join(map(lambda id: f"{id}" , row['Documents']))
    print(f"('{token}', ARRAY[{documents}]),")
    
```
